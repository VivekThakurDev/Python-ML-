```python
# import numbers as np
# import pandas as pd
# import matplotlib.pyplot as plt
# from learn.model_selection import train_test_split # type: ignore
# from sklearn.ensemble import  RandomForestClassifier
# from sklearn.metrics import accuracy_score
import joblib
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

```

```python
df=pd.read_csv("/content/sme_data.csv")
df
```

```python
x=df[["SepalLengthCm",	"SepalWidthCm",	"PetalLengthCm",	"PetalWidthCm"	]]
y=df[["Species"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
clf= RandomForestClassifier()
clf.fit(x_train,y_train)  # fit model
```

```python
#  After this mainly we use to check the ml algorithm
# prediction model

y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
y1 = clf.predict([[5.1,3.5,1.4,0.2]])
```
